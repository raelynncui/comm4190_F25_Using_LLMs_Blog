## Context

Recently, while watching a YouTube video, I noticed a new "Ask" button that appeared below the video. When I clicked it, a chat window popped up to the right of the video, where it prompted me to ask questions about the video I was watching. 

<img src="ai.png" width="80%"/>

I had never noticed this feature before, but it has apparently existed for a while -- previously only for Premium users. However, it seems like it's now available for all accounts over the age of 18, according to YouTube. 

Emotions online seem mixed -- some people are excited about this feature (since it would allow users new to a channel with lots of history to get caught up quicker, for example), whereas others are not (since it poses privacy concerns for these video creators). 

This new feature is just one of many AI-driven innovations that YouTube has released in [2024](https://blog.youtube/inside-youtube/2024-in-youtube-ai/). Others include:
- AI Video Summaries
- Live Chat Summaries
- Ask Music (users can describe a mood or vibe and get a curated playlist)
- Inspiration (helps content creators brainstorm ideas)
- Auto Dubbing
- Dream Screen (allows content creators to create AI-generated backgrounds for videos)
- Dream Track (allows content creators to create AI-generated soundtracks for videos)
- Posts (a space for content creators to workshop new posts with the help of AI)

## How does it work?

I couldn't find much online about what's powering this LLM. According to [YouTube Help](https://support.google.com/youtube/answer/14110396?hl=en#zippy=%2Cwho-can-use-the-conversational-ai-tool-on-youtube%2Chow-does-the-conversational-ai-tool-on-youtube-work%2Cwhat-data-is-collected-how-is-it-used):

> This tool is designed to provide answers to questions about videos you're watching. Responses are generated by large language models (LLMs). The LLMs draw on info from YouTube and the web.

Based on my brief experience with this tool, which I describe below, the LLM seems to make use of the video's transcript and timestamps to generate answers.

## My experience

In the video I was watching, the creator was making a banana cream latte in a very casual, vlog-style format (not a formal recipe vide). So, I prompted "give me the recipe she uses for the banana cream latte." I figure that this would be a good test for the LLM's capabilities, since the steps and ingredients for this recipe were scattered throughout the clip and not cleanly formatted. 

<img src="1.png" width="30%"/>

<img src="2.png" width="30%"/>

The response was formatted as a concise, easy-to-follow recipe with timestamps accompanying each step. The abundance of timestamps was really interesting to me, and I didn't expect it. This makes me think that the results are based on a time-stamped transcript, which the LLM reads and then generates a response based on.

Overall, this response was surprisingly impressive! I verified the recipe and the LLM was able to pinpoint each of the steps, despite the person in the video occassionally going off topic and discussing things other than the recipe. This shows that the LLM was able to capture the continuity of information across a longer period of time and was not thrown off by off-topic transcriptions. Additionally, the LLM picked up on the precise wording of the creator, incorporating the brand names correctly (Monin, Nescafe) as well as nuanced measurements ("tiniest splash"). 

## Reflection

As a YouTube viewer myself, I can't think of scenarios in which I would want to engage with this feature, as cool as it is. This is because watching videos is more of a passive activity for me -- I often leave YouTube videos on as I'm doing chores, eating, and getting ready for the day. During these activities, there's not really a time when I would want to drop what I'm doing and ask the LLM a question.

However, I can definitely see the potential of this feature for tutorial and educational videos. For example, if I was viewing a math video and was having trouble understanding the concept that the instructor was writing on the screen, this feature could defintely come in handy (however, this would require it to be able to visually analyze the frames of the video and not rely solely on the transcripts). I think it's an interesting strategic push by YouTube to make viewing more immersive and interactive, and I'm excited to see how AI changes the media landscape even more in the future!
