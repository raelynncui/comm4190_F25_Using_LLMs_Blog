# Book Reflection #4

In the chapter "AI as a Coworker," Mollick states that "The ability to outsource crappy, meaningless tasks to the AI can be freeing. The worst parts of your job go to AI so that you get to focus on the good stuff."

If I had seen this quote a few months ago, I think I would've completely agreed. However, after reflecting on how I've been using LLMs in my personal and academic spheres of life this past year, I don't think I would 100% be on board with this argument.

<img src="cover.jpeg" width="100%"/>

When I consider why I would've agreed to this statement a few months ago, I think it comes down to the fact that I'm able to use AI to perform (mainly academic) tasks that would've been extremely tedious and mentally exhausting to do manually. For example, if I was having trouble breaking down a math equation, I could have ChatGPT walk me through it instead of having to scour YouTube for an answer. If I didn't understand an error I got when I was programming, I could have an LLM assist me in debugging instead of having to search through online forums for others' solutions.

Going back to Mollick's quote, he says that "the worst parts of your job go to AI so that you get to focus on the good stuff." In the previous scenario I described, I think this is definitely true. However, over time, I think that what I've defined as the "good stuff" has changed, and this seems to be representative of many others' experiences, not just mine.

<img src="fulfillment.jpeg" width="60%"/>

By "good stuff," I'm imagining that Mollick is refering to activities that we find self-fulfillment in (whether that's activities outside of work or more interesting/exciting things within work). By using AI to allow us to focus more on this "good stuff," however, I think our standards of what's "good" is continually rising. I don't think this is bad, but I do think there's danger in this growing gap between what's *actually* good for us and what we *perceive* to be good.

For example, when I look around to other students' screens during my classes, I always see at least one person taking a problem/question and copying it directly into ChatGPT to get a response (and this is something I'm guilty of myself). In these scenarios, it seems like we *are* freeing up time to focus on better stuff -- by copying this question into ChatGPT and pasting its response elsewhere, we can convince ourselves that we can now use this time for more fulfilling activities. However, in this process, I think we're losing the value of what's actually good for us (in this example, that would be learning how to critically think about the problem and how to communicate it as an answer).  

<img src="chatgpt.png" width="60%"/>

Therefore, (from a mainly student and learning-oriented perspective), I think that the ability to use AI to outsource these so-called "meaningless" tasks comes with the condition that we're able to actually understand what's meaningless and what's good for us. Currently, I feel myself hitting more of a plateau with LLMs, as I'm beginning to use them to outsource work that *feels* tedious and meaningless for me to do, but actually isn't. 

It actually reminds me a bit of Plato's Allegory of the Cave -- where the outdoors is what's truly the "good stuff" capable with AI that I haven't yet realized as a student. Currently, I feel stuck inside the cave a bit -- relying on LLMs to assist with tasks that I think are freeing me up to do more fulfilling things, but in reality, aren't getting me anywhere far.
