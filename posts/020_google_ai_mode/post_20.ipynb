# Google AI Mode

<img src="cover.png" width="100%"/>

Recently, I was listening to Lenny Rachitsky's [podcast](https://www.lennysnewsletter.com/p/how-google-built-ai-mode-in-under-a-year) with Robby Stein, VP of Product at Google Search. In this podcast, they talked about the introduction of Google's AI Mode, a separate tab on the search page that leverages Google's search capabilities in a more conversational way.

For example, this is the responses to the prompt "what's the best running shoes to buy?"

<img src="sample.png" width="70%"/>

This is AI Mode's own description of itself:
> Comprehensive answers: This feature creates a comprehensive response by dividing complex queries into subtopics and searching for them simultaneously across the web, rather than just providing a list of links.

> Powered by Gemini 2.5: The feature uses the advanced reasoning and multimodal capabilities of Google's Gemini 2.5 model to deliver detailed responses to complex questions.

> Multimodal input: Users can input queries by typing, speaking, or uploading an image.

> Follow-up questions: The conversational nature allows follow-up questions to explore a topic further.

> Grounded in web content: Responses are based on high-quality web content to improve factual accuracy. They also include helpful links for further exploration.

> Integrated with other Google products: This feature can incorporate information from the Google ecosystem, including Maps, Reviews, and Shopping, for more detailed results.

> Deep Search capability: The most advanced research tool, Deep Search, can browse hundreds of sites to generate a comprehensive, cited report for Google AI subscribers. 


In this podcast, an interesting point was brought up -- previously, the "goal" with Google search was to take your query ("what's the best running shoes to buy?") but make it as concise and bare-bones as possible when actually typing it in the search bar ("best running shoes"). In fact, I distinctly remember learning about these "best practices" in elementary school. Our teachers would explain to us that it wasn't worth typing things like "what is" and "how do" into the search bar -- these words are unnecessary to the actual output of the search.

For example, Google even states this on their own website:

<img src="concise.png" width="70%"/>

However, with the rise in popularity of chat-based LLMs, we seem to be reversing and going in the opposite direction now -- replacing concise queries with lengthy, more descriptive prompts. 

<img src="long_prompt.jpg" width="70%"/>

I think this will mark a really interesting shift in how we converse with the search interface. It almost seems as if this new AI mode is encouraging people to revert back to their "bad" habits; however, these bad habits may actually be producing more optimal search results this time. For example, in the image above, the user would've likely gotten less specific and tailored results if they had just tyled "weekend Nashville activities." 

At the same time, I think this also poses challenges for user groups (like those from older generations) that have these more concise practices engrained in their behavior. For instance, they may stick to these short and direct prompts, never venturing to the more detailed and vivid prompts that yield even more impressive results. 

Overall, I think Google's AI Mode marks a new era in the search experience. With its own dedicated panel on the search page, I'm interested in seeing if people will adopt it in their habitual experiences going forward, or if they will stick to their old practices on the non-coversational interface.
