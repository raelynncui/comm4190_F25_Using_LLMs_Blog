## What is Atlas?

Last month, while using ChatGPT, a small pop-up appeared at the bottom of my chat window that suggested trying Atlas. Atlas is OpenAI’s new AI-powered web browser (launched in October 2025) that integrates ChatGPT directly into browsing so users can ask questions, summarize pages, and perform tasks without leaving the site they’re on. 

It has many features, such as:
- An in-page sidebar assistant
- Optional “browser memory” to recall context across sessions
- A premium “agent mode” that lets the AI take actions like filling forms or navigating websites

Overall, Atlas adds a conversational layer on top of browsing, reframing the browser as more of an an AI companion.

In this blog post, I'll talk about my initial experience with Atlas!

## Setting Atlas up

After I downloaded Atlas, the first step was choosing a Chrome profile that I wanted to sync with it. 

<img src="import_chrome.png" width="30%"/>

I even had the option to enable browser memory, which would give the LLM access to my search history. I thought this feature was really cool, since I can think of many uses cases for this/times when I've wanted to find something in my history, but the standard Chrome history search feature didn't work.

Personally, the idea of Atlas having access to my browsing history doesn't make me uncomforable, but I can image that it does for many others. Therefore, I'm glad that Atlas made this privacy option very clear, allowing users to easily opt out.

## The cold start screen

<img src="cold_start.png" width="70%"/>

Upon cold start (first opening the app, without entering anything), I realized that I had connected Atlas to the wrong Chrome profile (whoops!). Regardless, a few things stuck out to me -- first, the browser retrieved my Chrome bookmarks and placed them similarly at the top of the browser screen. Additionally, the queries that it recommended seemed to be partially influenced by my ChatGPT conversation history. I'm not sure where the UPS plane crash came from (maybe trending news, but I had never had a chat with ChatGPT about this), but the other suggestions -- "Internships by company" and "Compare PM interview prep" were definitely influenced by my chat history. These suggestions didn't seem to be based on my browser history, since I don't use this browser for internship/interview research.

In terms of UX, I really enjoyed the experience of opening this app. The interface is clean and not overwhelming, and the app even emitted some sort of color-changing glow across my desktop!

## Asking a question

For this experiment, I wanted Atlas to help me find tarnish-resistent jewelry from the brand *en route*.  

<img src="beginning.png" width="70%"/>

Similar to the ChatGPT interface, Atlas showed snippets of the model's through process. In this screenshot, you can see it "thinking":
> I'm thinking through how best to look for jewelry that won't tarnish on the en route jewelry website. I need to use my capabilities wisely to find the right pieces to meet the user's needs and do a good job. Let's get started on this task!

Already, this train of thought seemed very human-like and almost agentic. By stating that it has to "meet the user's needs and do a good job," it seems to acknowledge that it's doing work outside the realm of a simple chat companion. 

<img src="1.png" width="70%"/>

Then, it started searching online. You can see some search results that come up when it decides to search "En route jewelry won't tarnish pieces." Behind the scenes, I'm guessing that my initial prompt goes through a system that produces some optimal search query for Google, which is what's showing up here.

<img src="2.png" width="70%"/>
Here, it seems to be scanning the website for any mention of "tarnish-free."

This was the point where I think the LLM didn't handle the query optimally. I was hoping that it would take some steps to understand what would imply a tarnish-free product (researching different mental compositions, plating, etc.) and then search for those compositions across listings (since en route lists the materials for each product).

Instead, it seemed to just scan the site for any mention of "tarnish-free," not digging as deep as I had hoped.

<img src="3.png" width="70%"/>

## Output

For the final output, Atlas generated a correct list of products from the website that were tarnish-free. While it was able to identify these products, when I asked it to place these items in my cart, it wasn't able to. I was surprised at this, since the marketed agentic capabilities of Atlas made it seem like it should be able to.

However, this may be a specific issue with the en route website -- maybe adding something to a user's cart requires some sort of API on the backend that Atlas didn't have access too. This reminds me of some podcasts I've been listening to regarding the future of web search with the introduction of agents -- sites now have to be "API first" and consider not just how humans are interacting with the interface, but also agents. After playing around with Atlas, I can definitely see how important this is and I'm interested to see how this agent/site interactivity will have an impact on website interfaces and backends!
