# Can an LLM Predict my Spotify Wrapped?

<img src="cover.png" width="100%"/>

Recently, Spotify released Spotify Wrapped: an annual, personalized year-in-review feature where Spotify compiles and shares each user's listening habits. Users can see their top artists, top songs, top genres, total time listened, and so on. In this year's Wrapped, they even incorporated some new AI-powered insights -- the end of my wrapped had a set of "reports" that said "powered by AI" at the bottom: 

<img src="folder.png" width="20%"/>

In this blog, I want to see if LLMs can predict my Spotify Wrapped (top 5songs, top 5 artists, total minutes listened, and top genre) just based on demographics/characteristics non-related to music. I will prompt three LLMs (GPT-5.1, Sonnet 4.5, and Gemini) with the following prompt:

> I am a Spotify listener. I want you to predict my top 5 artists, top 5 songs, total minutes listened, and top genre of 2025 just based on my demographics/things about me not related to music. Ask me as many questions as you need to make an accurate prediction. Once I answer your questions, generate your prediction.


For context, this is my actual wrapped. Let's see how well these LLMs do!

# Reflection

While I was initially critical of how verbose and excessive ChatGPT seemed to be with its questions, after seeing the predictions of the more concise and direct models (Claude and Gemini), I think some of these questions actually could've been extremely valuable. Because Claude's and Gemini's predictions seemed to hinge around the stereotypical "20-year-old female," they failed to pick up on more nuanced aspects of my listening behavior. On the other hand, GPT-5.1 seemed to inquire more about my aesthetics and vibes -- if I had answered in more detail or answered in a different way, I'm wondering if it could've generated better predictions.
